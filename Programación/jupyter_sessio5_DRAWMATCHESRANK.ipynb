```python
#Mostrem el ranking de les 10 imatges que mes s'asemblen segons el sistema desenvolupat 
files = open((os.path.join(params['root'],params['database'],'val','result','2807-7096-24202'+'.txt')),'r')
i=0
print("Imatge a analitzar en el rank")
os.chdir('C:\Users\Gerard\Documents\Universidad\TercerCurso\GDSA\Proyecto\egara_view\TerrassaBuildings900\\val\images')
img=mpimg.imread('2807-7096-24202.jpg')
plt.imshow(img),plt.show()
bf = cv2.BFMatcher()
sift = cv2.SIFT()
#Buscamos los descriptores de la imagen de entrada
keypoints1, descriptors1 = sift.detectAndCompute(img,None)

os.chdir('C:\Users\Gerard\Documents\Universidad\TercerCurso\GDSA\Proyecto\egara_view\TerrassaBuildings900\\train\images')
print("Ara mostrem el ranking")
for i in range(0,10):
    img2=mpimg.imread(files.readline().replace('\n','')+'.jpg')
    i=i+1
    #Obtenemos los keypoints de la segunda imagen a analizar 
    keypoints2, descriptors2 = sift.detectAndCompute(img2,None)
    #obtenemos las coincidencias entre los descriptores de las imagenes a comparar
    matches = bf.match(descriptors1,descriptors2)
    #Los ordenamos
    matches = sorted(matches,key=lambda val: val.distance)
    #Tamaño de las coincidencias encontradas
    print len(matches)
    #reducimos a 20 coincidencias y las pintamos encima de cada foto unidas por lineas
    out = drawmatches(img, keypoints1,img2, keypoints2, matches[:20])
    #mostramos las imagenes
    imshow(out)
    show()
files.close()
```
