#### Obtenim les caracteristiques i realitzem el ranking basat en les distancies euclideas

La nostra funcio get_features inclou totes les creades per aquesta sessio: get_local_features, 
train_codebook, get_assignments i build_bow

Fem un resize de 250 per aquesta Notebook en comptes del que utilitzem normalment (800) per a que el codi tardi menys en executar-se ja que com volem treure a mes una grafica amb diferents valors del Codebook necessitarem executar-lo mes d'una vegada

Tambe en aquesta Notebook hem modificat la funcio get_features de forma que li podem passar el Codebook size i executar-lo amb diferents valors, pero aquesta modificacio la fem exclusivament per aquesta Notebook

```python
lista_paraules=list()
ap=list()
mAP=list()
lista_paraules=[100, 1000, 2000, 4000, 8000, 10000]
params['max_size'] = 250
for paraules in lista_paraules:
    get_features(params,paraules)
    #Calcul del ranking
    rank(params)
    #Evaluacio del ranking
    ap_list=eval_rankings(params)
    ap.append(ap_list)
    mAP.append(np.mean(ap_list))
```

Utilitzem diferents valors pel Codebook: 100, 1000, 2000, 4000, 8000 i 10000
