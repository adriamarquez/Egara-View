```python
#Leemos la imagen a procesar para obtener los Kpoints
img=cv2.imread('Mushroom2.png')
```

```python
#Una vez tenemos la imagen leida procesaremos una serie
#de funciones que extraeran los puntos de interés mediante
#un sistema basado en ORB.
#Habilitamos el módulo que nos permitirá utilizar el detector
orb=cv2.ORB()
#Ejecutamos la función, insertandole la imagen deseada y almacenamos los keyp.
kp=orb.detect(img,None)
#Los computamos y los ubicamos sobre la imagen
kp, des= orb.compute(img,kp)
```

```python
#A continuación se muestran los resultados, los puntos
#obtenidos son ubicados encima de la imagen original 
#para obtener una representación gráfica
#Dibujamos esos puntos en una imagen de salida llamada imagen 2 que será la suma
#de los puntos detectados más la imagen de entrada.
img2 = cv2.drawKeypoints(img,kp,color=(0,255,0), flags=0)
#Mostramos por pantalla
imgplot=plt.imshow(img2)
```
